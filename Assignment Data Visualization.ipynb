Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
np.random.seed(42) 
n_samples = 1000
x1 = np.random.normal(0, 1, n_samples)
x2 = x1 + np.random.normal(0, 0.5, n_samples)
x3 = x1 + x2 + np.random.normal(0, 0.2, n_samples)
```

```python
data = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3})
```

```python
correlation_matrix = data.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
```

Q1.  Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
```

```python
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Z')
fig.update_layout(title='3D Scatter Plot', scene=dict(
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    zaxis_title='Z Axis'
))
fig.show()
```

Q4.  Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(under root x^2 + y^2).

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
```

```python
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
```

```python
fig.update_layout(title='3D Surface Plot of z = sin(√(x² + y²))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))
fig.show()
```

Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource
from bokeh.transform import transform
```

```python
output_notebook()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the "sizes" and "colors" columns.

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 50, size=100)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=100)
```

```python
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))
```

```python
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y', background_fill_color="#fafafa")
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)
output_notebook()
show(p, notebook_handle=True)
```

Q5.  Create a Bokeh heatmap using the provided dataset.

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tooltips=[("Value", "@image")])
```

```python
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())
```

```python
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
output_notebook()
show(p, notebook_handle=True)
```
